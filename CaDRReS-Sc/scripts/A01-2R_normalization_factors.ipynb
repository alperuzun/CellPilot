```r
count <- read.csv("../preprocessed_data/scRNAseq/AS_et_al_tpm.csv", row.names = 1)
head(count)
```

```r
TMM <- calcNormFactors(count, method="TMM")
write.csv(TMM, "../preprocessed_data/scRNAseq/normalization_factor/AS_et_al_tmm_factor.csv")

RLE <- calcNormFactors(count, method="RLE")
write.csv(RLE, "../preprocessed_data/scRNAseq/normalization_factor/AS_et_al_rle_factor.csv")

UQ <- calcNormFactors(count, method="upperquartile")
write.csv(UQ, "../preprocessed_data/scRNAseq/normalization_factor/AS_et_al_uq_factor.csv")
```

```r
clist <- c("AP_et_al", "SP_et_al")
for (i in clist) {
    
    count <- read.csv(sprintf("../preprocessed_data/scRNAseq/%s_tpm.csv", i), row.names = 1)
    
    TMM <- calcNormFactors(count, method="TMM")
    write.csv(TMM, sprintf("../preprocessed_data/scRNAseq/normalization_factor/%s_tmm_factor.csv", i))

    RLE <- calcNormFactors(count, method="RLE")
    write.csv(RLE, sprintf("../preprocessed_data/scRNAseq/normalization_factor/%s_rle_factor.csv", i))
    
    UQ <- calcNormFactors(count, method="upperquartile")
    write.csv(UQ, sprintf("../preprocessed_data/scRNAseq/normalization_factor/%s_uq_factor.csv", i))   
}
```
